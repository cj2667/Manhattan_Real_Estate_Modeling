```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("home_eda_with_listinginfo.csv")
```

```python
data
```

```python
# Clean bds null
data = data[pd.notnull(data["bds"])]
# print(data.shape[0])

# Clean ba null
data = data[pd.notnull(data["ba"])]
# print(data.shape[0])

# Clean sqft null
data = data[pd.notnull(data["sqft"])]
# print(data.shape[0])

# Clean year_built null
data = data[pd.notnull(data["year_built"])]

# Filter sold_year to 2020
data = data[data["sold_year"] == 2020]
# And drop it
data = data.drop(["sold_year"], axis=1)
# print(data.shape[0])

# Get rid of price outlier
data = data[data["price"] != 10]

# Drop sold_day, sold_month
data = data.drop(["sold_day"], axis=1)
##########################################
data = data.drop(["sold_month"], axis=1) #
##########################################

# Drop pets and laundry
data = data.drop(["pets"], axis=1)
data = data.drop(["laundry"], axis=1)
# print(data.shape[0])

# Drop heating, cooling, flooring
data = data.drop(["heating"], axis=1)
data = data.drop(["cooling"], axis=1)
data = data.drop(["flooring"], axis=1)

# Drop most_recent_listing_date, most_recent_listing_price, sold_time, price_percent_change
data = data.drop('most_recent_listing_date', axis=1)
#######################################################
data = data.drop('most_recent_listing_price', axis=1) #
#######################################################
data = data.drop('sold_time', axis=1)
data = data.drop('price_percent_change', axis=1)

####################################################
# Drop listing_sold_time_diff                      #
data = data.drop('listing_sold_time_diff', axis=1) #
####################################################
```

```python
data.isnull().sum()
```

```python
# Reset index
data = data.reset_index(drop=True)

# Convert parking column
# list of options
parking_option_list = data["parking"].tolist()
output = set()
for comma_string in parking_option_list:
    row_list = comma_string.split(", ")
    for option in row_list:
        output.add(option)
# Create multi_dummy columns
for option in output:
    data[option] = [0] * data.shape[0]
    for i in range(data.shape[0]):
        row_list = data.loc[i, "parking"].split(", ")
        if option in row_list:
            data.loc[i, option] = 1
# drop parking column
data = data.drop(["parking"], axis=1)
```

```python
output
```

```python
# Convert Appliances column
data = data.replace('See more facts and features', np.nan)
data = data.replace('Other interior features', np.nan)
# Get rid of "Appliances included: "
index = "Appliances included: ".find(":")
for i in range(data.shape[0]):
    if pd.isnull(data.loc[i, "appliances"]):
        pass
    else:
        data.loc[i, "appliances"] = data.loc[i, "appliances"][index + 2:]

# Reset index
data = data.reset_index(drop=True)
# list of Appliances options
appliances_option_list = data["appliances"].tolist()
output_appliances = set()
for comma_string in appliances_option_list:
    if pd.isnull(comma_string):
        pass
    else:
        row_list = comma_string.split(", ")
        for option in row_list:
            output_appliances.add(option)
# Create multi_dummy columns
for option in output_appliances:
    data[option] = [0] * data.shape[0]
    for i in range(data.shape[0]):
        if pd.isnull(data.loc[i, "appliances"]):
            pass
        else:
            row_list = data.loc[i, "appliances"].split(", ")
            if option in row_list:
                data.loc[i, option] = 1
# drop appliances column
data = data.drop(["appliances"], axis=1)
```

```python
output_appliances
```

```python
data.isnull().sum()
```

```python
# Turn year_built into building_age
data["year_built"] = pd.to_numeric(data["year_built"])
building_age = 2020 - data["year_built"]
data["building_age"] = building_age
data = data.drop(["year_built"], axis=1)
```

```python
data.isnull().sum()
```

```python
# Turn type into dummy variables
dummy = pd.get_dummies(data["type"])
# print(dummy)
# Combine and drop type
data = pd.concat([data, dummy], axis=1)
# print(data.head())
data = data.drop(["type"], axis=1)
# print(data.head())
```

```python
data.isnull().sum()
```

```python
# Adjust zip_code
for i in range(data.shape[0]):
    data.loc[i, "zip_code"] = data.loc[i, "zip_code"][-5:]
# data["zip_code"] = pd.to_numeric(data["zip_code"])
# col_zip = sorted(data['zip_code'].unique())
dummy_zip_code = pd.get_dummies(data["zip_code"])

# # Convert the zipcode columns into one PCA
# sub_data = dummy_zip_code
# N_COMPONENTS = 1
# pca_model = PCA(n_components = N_COMPONENTS)
# pc_matrix = pca_model.fit_transform(sub_data)
# df_PCA = pd.DataFrame(pc_matrix, columns = ['PCA_zip_code'])
# Add columns
data = pd.concat([data, dummy_zip_code], axis =1)
# Drop original zip_code
data = data.drop('zip_code', axis =1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error as MSE
```

```python
train, test = train_test_split(data, test_size = 0.3)
x_train = train.iloc[0:,2:]
y_train = train[['price']]
x_test = test.iloc[0:,2:]
y_test = test[['price']]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_estimators': (10,30,50),
    'max_depth':(4,5,6,8,10,15),
    'min_samples_split': (2,4,8),
    'min_samples_leaf': (4,8,12,16)
}

model = GridSearchCV(RandomForestRegressor(), parameters, cv = 3, iid = False)
model.fit(x_train, np.ravel(y_train))
model.best_score_,model.best_params_
```

```python
#put the parameter into the random forest model
rfmodel = RandomForestRegressor(max_depth =5, min_samples_leaf =4, min_samples_split =2, n_estimators = 10)
rfmodel.fit(x_train, np.ravel(y_train))

print("Training R-Square",rfmodel.score(x_train,y_train))
print("Testing R-Square",rfmodel.score(x_test,y_test))
```

```python
rfmodel.fit(x_train, y_train)
y_pred = rfmodel.predict(x_test)
mse_dt = MSE(y_test, y_pred)
rmse_dt = mse_dt**(1/2)
print("Test set RMSE of dt: {:.2f}".format(rmse_dt))

data["price"].describe()
```

```python
feature_names = [key for key in data if not key == 'price']

#visualize the feature importance in the random forest model
import matplotlib.pyplot as plt
%matplotlib inline
importances = rfmodel.feature_importances_
indices = np.argsort(importances)
plt.rc('ytick', labelsize=9)
plt.figure(figsize=(12,12),dpi=300)
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)),feature_names)
plt.xlabel('Relative Importance')
```
